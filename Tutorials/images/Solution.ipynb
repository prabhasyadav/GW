**Q5.   Effective Conductivity** (ca. 7 pts.)

Two confined aquifers are separated by an impervious layer (see figure). A withdrawal well is screened over the entire thicknesses ($m_1 = 4$ m, $m_2 = 2$ m) of both aquifers. Its total pumping rate $Q$ is $4·10^{-3}$ m³/s. Hydraulic conductivities of the two aquifers are equal to $K_1 = 1·10^{-5}$ m/s and $K_2 = 2.5·10^{-4}$ m/s, resp. 

<img src="figs/Q5.png" style="width: 200px;"/>

a.	Determine the effective horizontal hydraulic conductivity of the two-aquifer system.   (ca. 3 pts.)
(Hint: $mK = m_1K_1+m_2K_2$ or $m/K = m_1/K_1 + m_2/K_2$ ; choose the correct one for your answer)

b. Calculate discharges Q1 and Q2 withdrawn from the two aquifers. (Hint: Aquifer has a unit width)  (ca. 4 pts.)

Solution 5a - (L06/10)

The effective horizontal conductivity is given by:

$$
K_{eff_h} = \sum_{i=1}^n\frac{m_iK_i}{m}
$$
with $m = \sum m_i$ for $i = 1\cdots n$.

```python
#Given
m1, m2, K1, K2, Q = 4, 2, 1e-5, 2.5e-4, 4e-3

#Solution
m = m1+m2
Keff_h = (m1*K1+m2*K2)/m
print('The effective horizontal conductivity  is '  '%e'  %Keff_h, 'm/s') 
```

The discharge of the homogeneous layer is:
$$
Q = w\, m\, K_{eff_h}\Delta h/L
$$

with $w$ (= 1 considered here) the width of the aquifer and $L$ is the length of the flow along the layer. 

The discharges in each levels is:
$$
Q_i = w\,m_i\,K_i \Delta h/L
$$

with $i = 1 \text{ and } 2$. To obtain the discharge at each level we need to find $\Delta h/L$ from eq. 3 and then use it to obtain discharge $Q$ at each level using eq. 4.

```python
deltah_L = Q/(m*Keff_h)
deltah_L 
```

```python
Q1 = m1*K1*deltah_L
Q2 = m2*K2*deltah_L
print(u'The discharge at the top of the aquifer Q\u2081 is '  '%e'  %Q1, 'm/s') 
print(u'The discharge at the top of the aquifer Q\u2082 is '  '%e'  %Q2, 'm/s') 
print(u'Check Q\u2081 + Q\u2082  is '  '%e'  %(Q1+Q2), 'm/s') 
```

Solution 8a - (L10/10)

Solution 8b - (L10/20)

The first order decay model is given as:

$$
C(t) = C_0\exp^{-\lambda t}
$$
Rearranging it becomes
$$
\frac{C(t)}{C_0} = \exp^{-\lambda t}
$$
Taking ln both side we get
$$
\ln\frac{C(t)}{C_0} = -\lambda t 
$$
Rearranging and letting -ve sign go, will lead to
$$
\lambda = \frac{1}{t}\ln\frac{C_0}{C(t)}
$$

```python
# Given are
c0, ct, t = 100, 10, 10 

# Solve
lambd_a = 1/t*(np.log(c0)-np.log(ct))
T_half = np.log(2)/lambd_a

print('The decay constant is ' '%.3f' %lambd_a, '1/y' )
print('The half life of the massis ' '%.3f' %T_half, 'y' )
```
